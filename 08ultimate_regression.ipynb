We will evaluate our regression models on a given dataset and will find out which is best

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import the dataset

```python
data=pd.read_csv('./datasets/Ultimate_Data.csv')
data.head()
```

Splitting into dependent and independent variables

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
x
```

```python
y
```

Splitting the data into x train and y train

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

Training the Multiple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
linear_regressor=LinearRegression()
linear_regressor.fit(x_train,y_train)
```

Predicting test results

```python
y_pred = linear_regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))   #ek saath dekh sake
```

Let's evaluate the model based on r2 score

```python
from sklearn.metrics import r2_score
print(f"The r2 score of the model is {r2_score(y_test,y_pred)*100}")
#93.01 percentage
```

Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=4)  #polynomial regressor of degree 4
x_poly=pf.fit_transform(x_train)  #converting into required matrix
print(x_poly) 
```

Train the model

```python
polynomial_regressor=LinearRegression()
polynomial_regressor.fit(x_poly,y_train)
```

Prediction of the test results 
firstly convert the x_test into required matrix of poly_x

```python
y_pred=polynomial_regressor.predict(pf.transform(x_test))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Let's evaluate this model

```python
print(f"The r2 score of this model is {r2_score(y_test,y_pred)*100}")
#94.31 percentage better than multiple linear regression
```

Support Vector Regression

In support vector we have to convert the y into 2d before the splitting into train and test or we can do it later also

```python
y_train=y_train.reshape(len(y_train),1)
y_test=y_test.reshape(len(y_test),1)
```

Feature Scaling is needed in support vector

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
sc_y=StandardScaler()
y_train=sc_y.fit_transform(y_train)
```

Train the model

```python
from sklearn.svm import SVR
support_vector_regressor=SVR(kernel='rbf')
support_vector_regressor.fit(x_train,y_train)
```

Prediction of test set results

```python
y_pred=sc_y.inverse_transform(support_vector_regressor.predict(sc_x.transform(x_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

#hamne y_pred ko inverse transfor kr k liya hai and  y_test ko kabhi transform hi nhi kiya
```

Evaluate the model

```python
print(f"The r2 score of support vector regression is {r2_score(y_test,y_pred)*100}")
#94.68 percent
```

Decision Tree Regression

we have to again split into x_train due to use of scaler in previous model

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

Training the model

```python
from sklearn.tree import DecisionTreeRegressor
decision_regressor=DecisionTreeRegressor(random_state=42)
decision_regressor.fit(x_train,y_train)
```

Prediction of test results

```python
y_pred=decision_regressor.predict(x_test)
np.printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Evaluation of model

```python
print(f"The r2 score of the model is {r2_score(y_test,y_pred)*100}")
#93.27 percent
```

Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_regressor=RandomForestRegressor(n_estimators=100,random_state=42)
random_forest_regressor.fit(x_train,y_train)
```

Prediction of Test Set

```python
y_pred=random_forest_regressor.predict(x_test)
np.printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

Evaluation of model

```python
print(f"The r2 score of the random forest is {r2_score(y_test,y_pred)*100}")
#95 percent
```

so for this dataset random forest is best
