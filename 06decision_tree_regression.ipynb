Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
Data=pd.read_csv('./datasets/Position_Salaries.csv')
Data.head()
```

No need of feature scaling in Decision Trees and Random Forest

Dividing into dependent and independent variables

```python
x=Data.iloc[:,1:-1].values
y=Data.iloc[:,-1].values
x
```

```python
y
```

Train the model

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
```

Let's predict one output at 9

```python
prediction=regressor.predict([[9]])
prediction  #pura 50k
```

Visualisation :- Not so necessary as when many features are present there is no use of visualisation

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Position vs Salaries (Deciesion Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
