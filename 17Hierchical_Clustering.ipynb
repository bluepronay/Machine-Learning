It is second part of clustering here we are doing Hierchical Clustering
here also we are doing aglomotive Clustering

1. Consider each datapoints as single cluster
2. then add two nearest clusters to form a single cluster  now n-1 clusters
3. then again two nearest clusters to form another cluster  now n-2 clusters present
4. See the dendrogram to determine the optimal no of clusters

Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing the dataset

```python
data=pd.read_csv('./datasets/Mall_Customers.csv')
data.head(3)
```

```python
x=data.iloc[:,[3,4]].values
x
```

For Dendrograms

```python
from scipy.cluster import hierarchy as sch
```

```python
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))  
#linkage linking 2 arguments, x and most recommened clustering technique
plt.title('Dendrogram')
plt.xlabel('No of Clusters')
plt.ylabel('Euclidean distances')
```

5 clusters , if we cut the longest lines

Train the model

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = hc.fit_predict(x)
print(y_hc)
```

```python
plt.scatter(x[y_hc == 0, 0], x[y_hc == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_hc == 1, 0], x[y_hc == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_hc == 2, 0], x[y_hc == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_hc == 3, 0], x[y_hc == 3, 1], s = 50, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_hc == 4, 0], x[y_hc == 4, 1], s = 50, c = 'magenta', label = 'Cluster 5')

# # Plot the centroids
# centroids = hc.cluster_centers_
# plt.scatter(centroids[:, 0], centroids[:, 1], s = 100, c = 'yellow', label = 'Centroids')
#no concept of centroids in hierchical clustering

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
